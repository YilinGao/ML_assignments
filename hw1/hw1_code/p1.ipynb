```python
import matplotlib.pyplot as plt
```

```python
plt.plot([0,1,1], [1,0,1], 'b+', label=r'$y=1$')
plt.plot([0], [0], 'bo', label=r'$y=0$')
plt.plot([-0.1, 0.6], [0.6, -0.1], 'r-')
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.text(0.4, 0.18, r'$f(x_1, x_2) = x_1 + x_2 - 0.5$', fontsize=12)
plt.axis([-0.1, 1.1, -0.1, 1.1])
plt.legend(loc='right')
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([0, 1, 0, 1], [0, 0, 0, 0], [0, 0, 1, 1], 'b+', label=r'$y=1$')
ax.scatter([0, 1, 0, 1], [1, 1, 1, 1], [0, 0, 1, 1], 'bo', label=r'$y=0$')
surface_x = np.arange(0, 1, 0.25)
surface_z = np.arange(0, 1, 0.25)
surface_x, surface_z = np.meshgrid(surface_x, surface_z)
surface_y = 0.5
ax.plot_surface(surface_x, surface_y, surface_z)
ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.set_zlabel(r'$x_3$')
ax.text(0.5, 0.4, 0.9, r'$f(x_1, x_2, x_3) = x_2 - 0.5$', fontsize=12)
plt.legend(loc='right')
plt.show()
```
